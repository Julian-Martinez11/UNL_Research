# Pair names

The goal of this notebook is to create tools to find the name of each of the 561 pairs.

### Pairs using the names from json file

```python
names_json = ["SM","cQu8","ctu8","cQu1","ctd8","cQlMi","cpQ3","ctlTi","ctG","ctZ","ctW","ctli","ctb8","cbW","cQl3i","cQq13",
              "cptb","ctp","ctei","cpQM","ctlSi", "cQq83","cQq81","ctq1","ctu1","cQei","cQb8","cpt","ctq8","cQd1","cQq11",
              "cQd8","ctd1"]

# All 561 combinations of coefficients (i.e SM,SM; SM,cQu8; cQu8,cQu8; SM,ctu8; cQu8,ctu8; ...)
name_combinations = []
for i in range(len(names_json)):
    for a in range(i+1):
        name_combinations.append([names_json[i],names_json[a]])
```

### Pairs using easier names like SM, EFT1, EFT2, etc.

```python
easy_names = ['SM']

for i in range(1,33):
    easy_names.append(f'EFT{i}')

# All 561 combinations of coefficients using easy names (i.e SM,SM; EFT1,SM; EFT1,EFT1; ...)
easy_name_combinations = []

for i in range(len(easy_names)):
    for a in range(i+1):
        easy_name_combinations.append([easy_names[i],easy_names[a]])
```

### The following code helps us to find what index corresponds to a pair using their easy name

```python
# Let's find what indexes correspond to which pairs of coefficients. 

# First, put the pair of coefficients in c1 and c2 in whatever order.
c1 = 'EFT1'
c2 = 'SM'

# Look for both possible combinations of the pair and print index:
if ([c1,c2] in easy_name_combinations) == True:
    print(f'The Index for the pair {[c1,c2]} is {easy_name_combinations.index([c1,c2])}')
else:
    print(f'The Index for the pair {[c1,c2]} is {easy_name_combinations.index([c2,c1])}')
```

### The following code helps us to create a dictionary that assigns:
* Keys: Each of the 33 Coefficients being represented by their easy name.
* Values: Each index at which the coefficient appears (i.e SM appears in index 0, index 1, index 3, ...)

```python
names_dictionary = {}

# Finding where the coefficients appear:
for i, json in enumerate(names_json):                                         # Enumerating names from json.
    for n, name in enumerate(name_combinations):                              # Enumerating all the combinations.
        if json in name:                                                      # If the name from json is in all combinations:
            names_dictionary.setdefault(easy_names[i],[]).append(n)           # add its easy corresponding name to the dictionary and 
                                                                              # where it appears (index)
```
